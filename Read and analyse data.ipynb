```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def file_to_data(filename):
    df = pd.read_csv(filename, usecols=[4,10,11], skiprows = 5099997, nrows = 1).to_numpy()[0]
    return df
```

## Plot: delta lambda from G 0,5 GHz to 3,5 for small powers

```python
def find_dlam(G):
    p_in = np.arange(0.05,0.35,0.05)
    theta_st, T_st, R_st = np.zeros(len(p_in)),np.zeros(len(p_in)),np.zeros(len(p_in))
    A = np.sqrt(p_in)
    for i,A_in in enumerate (A):
        G_in = file_to_data(f"Gext = {G}/p_in = {A_in**2}")
        theta = G_in[0]
        T = G_in[1]
        R = G_in[2]
        theta_st[i] = np.array(complex(theta).real)
        T_st[i] = np.array(complex(T).real)
        R_st[i] = np.array(complex(R).real)
    c = 3*10**8 #nm/ns
    lam0 = 1555
    w0 = 2*np.pi*c/(lam0)
    betha = 4e-6
    delta_w = w0*betha*theta_st
    delta_lam = lam0**2*delta_w/(2*np.pi*c)
    print(f"For Gext = {G}"+
            f"\n theta_st = {theta_st}"+
            f"\n T_st = {T_st}"+f"\n R_st = {R_st}"+f"\n dlam = {delta_lam}")
    return theta_st, T_st, R_st, delta_lam
```

```python
theta_stG05, T_stG05, R_stG05, dlam_G05 = find_dlam(0.5)
```

```python
theta_stG1, T_stG1, R_stG1, dlam_G1 = find_dlam(1.0)
```

```python
theta_stG15, T_stG15, R_stG15, dlam_G15 = find_dlam(1.5)
theta_stG2, T_stG2, R_stG2, dlam_G2 = find_dlam(2.0)
theta_stG25, T_stG25, R_stG25, dlam_G25 = find_dlam(2.5)
theta_stG3, T_stG3, R_stG3, dlam_G3 = find_dlam(3.0)
```

```python
theta_stG4, T_stG4, R_stG4, dlam_G4 = find_dlam(4)
theta_stG5, T_stG5, R_stG5, dlam_G5 = find_dlam(5)
theta_stG6, T_stG6, R_stG6, dlam_G6 = find_dlam(6)
theta_stG7, T_stG7, R_stG7, dlam_G7 = find_dlam(7)
theta_stG8, T_stG8, R_stG8, dlam_G8 = find_dlam(8)
```

```python
dlam_list = np.array([dlam_G05[-1],dlam_G1[-1],dlam_G15[-1],dlam_G2[-1],dlam_G25[-1],dlam_G3[-1],dlam_G4[-1],dlam_G5[-1],dlam_G6[-1],dlam_G7[-1],dlam_G8[-1]])
theta_list = np.array([theta_stG05[-1],theta_stG1[-1],theta_stG15[-1],theta_stG2[-1],theta_stG25[-1],theta_stG3[-1],theta_stG4[-1],theta_stG5[-1],theta_stG6[-1],theta_stG7[-1],theta_stG8[-1]])

Gext = np.array([0.5,1,1.5,2,2.5,3,4,5,6,7,8])
fig, ax1 = plt.subplots(dpi = 700) 
plt.title(r"$\Delta$ $\lambda$ and $\Theta$ over $_{ext}$")
ax1.set_xlabel(r'$_{ext}, GHz$') 
ax1.set_ylabel(r'$\Delta$$\lambda$, nm', color = 'green') 
ax1.plot(Gext, dlam_list, color = 'green') 
ax1.tick_params(axis ='y', labelcolor = 'green') 
  
# Adding Twin Axes

ax2 = ax1.twinx() 
  
ax2.set_ylabel(r'$\Theta, K$', color = 'blue') 
ax2.plot(Gext, theta_list, '*',color = 'blue') 
ax2.tick_params(axis ='y', labelcolor = 'blue') 
ax1.grid()
plt.show()
```

## For 0,35 mW T and R vs Gext:

```python
T_list = np.array([T_stG05[-1],T_stG1[-1],T_stG15[-1],T_stG2[-1],T_stG25[-1],T_stG3[-1],T_stG4[-1],T_stG5[-1],T_stG6[-1],T_stG7[-1],T_stG8[-1]])
R_list = np.array([R_stG05[-1],R_stG1[-1],R_stG15[-1],R_stG2[-1],R_stG25[-1],R_stG3[-1],R_stG4[-1],R_stG5[-1],R_stG6[-1],R_stG7[-1],R_stG8[-1]])
T_R = T_list+R_list
```

```python
plt.figure(figsize=(6, 4),dpi = (700))
plt.title(r"T and R over $_{ext}$")
plt.xlabel(r'$_{ext}, GHz$')
plt.grid()
Gext = np.array([0.5,1,1.5,2,2.5,3,4,5,6,7,8])
plt.plot(Gext, T_list, **{'marker': 'o'}, label = 'T')
plt.plot(Gext, R_list, **{'marker': 'o'}, label = 'R')
plt.plot(Gext, T_R, **{'marker': 'o'}, label = 'T+R')
plt.legend()
plt.show()
```
